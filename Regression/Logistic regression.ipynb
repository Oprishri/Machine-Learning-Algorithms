## Hard coded

```python
def sigmoid(x,t):
    return(1/(1+np.exp(np.negative(x.dot(t.T)))))
```

```python
def regression(n):
    m=train.shape[0]
    theta=[0]*x.shape[1]
    theta=np.matrix(theta)
    j=[]
    for i in range(n):
        l=sigmoid(x,theta)
        j.append(1/(2*m)*(-np.sum(y.T.dot(np.log(l))+(1-y).T.dot(np.log(1-l)))))
        if((j[i])<0.00001):
            break
        else:
            theta=gradient(theta.T)
    return(theta)
```

```python
def gradient(theta):
    l_rate=0.01
    theta=theta+l_rate*1/x.shape[0]*(x.T.dot(y-sigmoid(x,theta.T)))
    return(theta.T)
```

```python
def predict(n):
    sol=[]
    theta=regression(n)
    probability=sigmoid(test_x,theta)
    for i in range(len(probability)):
        if(probability[i]>0.5):
            sol.append(1)
        else:
            sol.append(0)
    sol=np.array(sol)
    return(sol)
```
