```python
data_train=np.matrix(train_data)
data_test=np.matrix(test_data)
m = data_train.shape[0]
v = np.ones((data_train.shape[0], 1))
x =  data_train[:,0:4]
y = np.append(v,x,axis=1).T
actual_value = data_train[:,4:5]
```

```python
n = data_test.shape[0]
v1 = np.ones((data_test.shape[0], 1))
x1 =  data_test[:,0:4]
y1 = np.append(v1,x1,axis=1).T
actual_value_test = data_test[:,4:5]
```

```python
alpha=0.001
alpha1=0.5
lambda1=0.01
```

```python
def elastic(epoach):
    error = []
    theta=np.matrix(np.ones([y.shape[0]]))
    for i in range(n):
        y_predicted = (theta.dot(y)).T
        J=(0.5/m)*(np.sum(np.square(y_predicted - actual_value))
                   +lambda1*((1-alpha1)*0.5*(np.sum(np.square(theta)))
                             +(alpha1*np.sum(theta))))
        error.append(J)
        if error[i] < 0.01:
             break
        else:
             theta = gradient_descent(theta,y_predicted)
    return(theta,error)
```

```python
def gradient_descent(theta,y_predicted):
    alpha=0.0000001
    theta = theta - (alpha/m)*(((y.dot(y_predicted - actual_value)).T)
                               +lambda1*((1-alpha1)*theta+alpha1))
    return(theta)
```

```python
def sol(epoch):
    theta=elastic(epoch)[0]
    y1_predicted = (theta.dot(y1)).T
    return(y1_predicted)
```

```python
def rms(epoch):
   y1_predicted=sol(epoch)
   return(math.sqrt(np.sum(np.square(y1_predicted-actual_value_test))/n))
   
```

```python
def plot(epoch):
    error=elastic(epoch)[1]
    print("Plot between error and number of iteration is :")
    plt.plot([i for i in range(len(error))],error)
```
