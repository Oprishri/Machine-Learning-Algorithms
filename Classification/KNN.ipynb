```python
data1=data.values
```

```python
data2=data1[:,0:4]
```

```python
def KNN(k,l):
    e_d=[]
    for i in range(data2.shape[0]):
        e_d1=np.sqrt(np.sum(np.square(data2[i:i+1,]-l)))
        e_d.append(e_d1)
    e_d=(np.matrix(e_d)).T
    val=np.hstack([np.array(e_d),data1[:,4:5]])
    val=val[np.argsort(val[:,0])]
    k_1=val[0:k,:]
    if np.count_nonzero(k_1[:,1:2]=='setosa')>=(k/2):
        print('setosa')
    elif np.count_nonzero(k_1[:,1:2]=='virginica')>=(k/2):
        print('virginica')
    else:
        print('versicolor')
    return   
```

```python
KNN(5,[1.2,4,1.5,0.7])
```
